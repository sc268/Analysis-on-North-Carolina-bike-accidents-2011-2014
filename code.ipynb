```python
import pandas as pd
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="darkgrid")
```

```python
#dat = pd.read_json('north_carolina_bicycle_crash_data_heatmap_.json')
dat = pd.read_csv('nc_bike_accidents.csv', index_col = False)
dat['Crash_Date'] = pd.to_datetime(dat['Crash_Date'])
dat['year'], dat['month'] = dat['Crash_Date'].apply(lambda x: x.year), dat['Crash_Date'].apply(lambda x: x.month)
#dat['YearMonth'] = pd.to_datetime(dat['year'])
```

```python
#time series
tmp = dat.Crash_Date
month =  30*60*60*24*1e9
idx_rounded = pd.DatetimeIndex((tmp.astype(np.int64) // month + 1 ) * month)
ts = pd.Series([1]*len(dat), index = idx_rounded)
ts_raw_aggregate = ts.groupby(ts.index).sum()
ts_raw_aggregate.plot()
plt.xlabel('date')
plt.ylabel('number of accidents')
```
